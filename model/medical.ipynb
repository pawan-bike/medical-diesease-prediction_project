```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Linear regression model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
print("test score",lr.score(x_test,y_test))
print("test score",lr.score(x_train,y_train))
```

# SVM(svr) model

```python
from sklearn.svm import SVC,SVR
svr=SVR()
svr.fit(x_train,y_train)
```

```python
print("test score",svr.score(x_test,y_test))
print("train score",svr.score(x_train,y_train))
```

# Decissiontree Classifier model

# decissiontree classifier model with parameters

```python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
dtcwp=DecisionTreeClassifier(criterion='entropy',min_samples_split=3,min_samples_leaf=2)
dtcwp.fit(x_train,y_train)
```
